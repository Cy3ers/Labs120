```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('imgg.jpg', cv2.IMREAD_COLOR) 
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray)
```

```python
edges = cv2.Canny(gray, 50, 50)
```

```python
plt.imshow(edges)
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/180, minLineLength=10, maxLineGap=10,threshold=130)
```

```python
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (255, 0, 0), 1)
```

```python
plt.imshow(img)
```
